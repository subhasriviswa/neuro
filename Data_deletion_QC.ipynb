```python
import os
import pandas as pd

# read the CSV file and create a new_file_path column
dfa = pd.read_csv("/Users/subhasriviswanathan/Downloads/Untitled spreadsheet - T1.csv")
dfa['new_file_path'] = dfa['Participant ID'].apply(lambda x: f'MIDT_{x}_events.mat')

# specify the directory where the files are located
directory = "/Users/subhasriviswanathan/Downloads/V1_eventsdata"

# get the list of files to remove based on the "Functional Scan" column
files_to_remove = dfa[dfa['Functional Scan'] == "removed"]['new_file_path'].tolist()

# loop through the files and remove them one by one
for file_name in files_to_remove:
    file_path = os.path.join(directory, file_name)
    print(f"Attempting to remove file: {file_path}")
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"{file_name} has been removed.")
    else:
        print(f"{file_name} does not exist in the directory.")
```
